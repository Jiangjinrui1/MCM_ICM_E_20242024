```python
# 筛选需要信息(时间、严重性、地点)

def filter_data(df):
    filter_data=df[['Severity','StartTime(UTC)','City']]
    return filter_data
Abingdon_filter=filter_data(Abingdon)
Santa_Cruz_filter=filter_data(Santa_Cruz)
dates=Abingdon_filter['StartTime(UTC)']
Abingdon_filter
```

```python
# 去掉Severity为UNK的行
def drop_unknown(df):
    df=df[df['Severity'] != 'UNK']
    df.reset_index(drop=True,inplace=True)
    return df
Abingdon_filter=drop_unknown(Abingdon_filter)
Santa_Cruz_filter=drop_unknown(Santa_Cruz_filter)
Abingdon_filter
```

```python
# 获取Severity字典
def get_Severity_dic(df):
    event_types = df['Severity'].unique()
    dictionary = pd.DataFrame(columns=['Severity_DEGREE'])
    dictionary['Severity_DEGREE'] = event_types
    dictionary.to_csv('{}.csv'.format('dictionary_2'),index=False)
    return dictionary
ab_severity_dic=get_Severity_dic(Abingdon_filter)
ab_severity_dic

```

```python
# 获取Severity字典
def get_severity_dict(df):
    severity_dict={}
    for i in range(len(df)):
        severity=df.iloc[i,0]
        if severity in severity_dict:
            severity_dict[severity]+=1
        else:
            severity_dict[severity]=1
    return severity_dict
Abingdon_severity_dict=get_severity_dict(Abingdon_filter)
Santa_Cruz_severity_dict=get_severity_dict(Santa_Cruz_filter)
Abingdon_severity_dict
```

```python
# 为Severity创建等级映射字典
# 创建一个字典来映射自然灾害事件到等级
severity_levels = {
 
    'Light':2,
    'Moderate':5,
    'Heavy':7,
    'Severe':9
}
# 将字典映射到dataframe中
Abingdon_filter['Severity_DEGREE'] = Abingdon_filter['Severity'].map(severity_levels)
Santa_Cruz_filter['Severity_DEGREE'] = Santa_Cruz_filter['Severity'].map(severity_levels)
Abingdon_filter
```

```python
Santa_Cruz_filter
```

```python
# 统计地区发生各种Severity类型的次数
def get_severity_count(df):
    severity_count={}
    for i in range(len(df)):
        severity=df.iloc[i,0]
        if severity in severity_count:
            severity_count[severity]+=1
        else:
            severity_count[severity]=1
    df['Times']=df['Severity'].map(severity_count)
    #去掉Severity重复的行
    df=df.drop_duplicates(subset=['Severity'],keep='first')
    df=df.drop(['StartTime(UTC)'],axis=1)
    return df
Abingdon_severity_count=get_severity_count(Abingdon_filter)
Santa_Cruz_severity_count=get_severity_count(Santa_Cruz_filter)
Abingdon_severity_count
```

### 统计发生的频率


```python
# 按统计2021-2023各州发生各EVENT的rate

#统计观测时间长度
latest_date = pd.to_datetime(dates,format="%d-%b-%y %H:%M:%S").max()
early_date=pd.to_datetime(dates,format="%d-%b-%y %H:%M:%S").min()
row_count_Abingdon = len(Abingdon_filter)

# days_since_2000 = (latest_date - early_date).days

print("最近观测日期:", latest_date)
print("最远观测日期：",early_date)
# print("观测时间长度为:", days_since_2000)

# 计算观测时间内发生事件次数
# 2. 将每个组的"COUNT"总和除以days_since_2000
grouped = Abingdon_severity_count.groupby(['Severity'])["Times"].sum()
rate_Abingdon = grouped / row_count_Abingdon
grouped = grouped.reset_index()
# 3. 将结果保存到名为"rate"的变量中
rate_Abingdon = rate_Abingdon.reset_index()  # 将索引重置为列
rate_Abingdon.columns = ['Severity', "rate_Abingdon"]  # 重命名列
rate_Abingdon.to_csv("result/rate_Abington.csv", index=False)

rate_Abingdon
```

```python
row_count_Santa_Cruz = len(Santa_Cruz_filter)
grouped = Santa_Cruz_severity_count.groupby(['Severity'])["Times"].sum()
rate_Santa = grouped / row_count_Santa_Cruz
grouped = grouped.reset_index()
# 3. 将结果保存到名为"rate"的变量中
rate_Santa = rate_Santa.reset_index()  # 将索引重置为列
rate_Santa.columns = ['Severity', "rate_Santa"]  # 重命名列
rate_Santa.reset_index(drop=True, inplace=True)
rate_Santa.to_csv("result/rate_Santa.csv", index=False)
rate_Santa
```

```python
#计算风险因子
# 整合事件发生频率与强度
rate_Abingdon['Severity_DEGREE']=rate_Abingdon['Severity'].map(severity_levels)
rate_Santa['Severity_DEGREE']=rate_Santa['Severity'].map(severity_levels)
rate_Abingdon
```

```python
# 风险因子=rate*Degree之和

rate_Abingdon['Risk_Factor']=rate_Abingdon['rate_Abingdon']*rate_Abingdon['Severity_DEGREE']
rate_Santa['Risk_Factor']=rate_Santa['rate_Santa']*rate_Santa['Severity_DEGREE']
rate_Abingdon
result_Abingdon =rate_Abingdon['Risk_Factor'].sum()
result_Santa =rate_Santa['Risk_Factor'].sum()

# 打印结果
print("平均风险因子计算结果为：", result_Abingdon)
```

```python
# 统计每个年/月的各Severity发生次数
def get_month_severity_count(df):
    import pandas as pd

    # 假设df是你的DataFrame，包含'StartTime(UTC)'和'Severity'等列

    # 将'StartTime(UTC)'列转换为日期时间类型
    df['StartTime(UTC)'] = pd.to_datetime(df['StartTime(UTC)'], format='%Y/%m/%d %H:%M')
    # 添加'YearMonth'列，将年份和月份整合到一起
    df['YearMonth'] = df['StartTime(UTC)'].dt.to_period('M')
    # 按'YearMonth'和'Severity'列分组，计算每个组合的计数
    severity_counts = df.groupby(['YearMonth', 'Severity']).size().unstack(fill_value=0)
    

    return severity_counts

Abingdon_month_severity_count=get_month_severity_count(Abingdon_filter)
Santa_Cruz_month_severity_count=get_month_severity_count(Santa_Cruz_filter)
Abingdon_month_severity_count
```
