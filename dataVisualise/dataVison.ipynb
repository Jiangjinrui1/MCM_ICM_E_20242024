```python
def loadDataSet(fileName):
    data = pd.read_csv(fileName)
    return data
# data_dic=r"D:\VSCodeWorkSpace\2024_C\data\data_dictionary.csv"
# data_wim=r"D:\VSCodeWorkSpace\2024_C\data\Wimbledon_featured_matches.csv"
data_USA=r"D:\VSCodeWorkSpace\2024_C\data\StormEvents_details-ftp_v1.0_d2023_c20240117.csv"
# data_D=loadDataSet(data_dic)
# data_W=loadDataSet(data_wim)
data_A=loadDataSet(data_USA)
data_Test=loadDataSet(data_USA)
```

```python
data_A=data_A[['YEAR','MONTH_NAME','EVENT_TYPE','STATE']]
data_A['EVENT_TYPE']=data_A['EVENT_TYPE'].str.lower()
# 按STATE升序,EVENT_TYPE升序排序
df_sorted = data_A.sort_values(['STATE','EVENT_TYPE']).reset_index(drop=True)
df_sorted
```

```python
grouped = df_sorted.groupby(['YEAR', 'STATE', 'EVENT_TYPE'])
counts = grouped.size().reset_index(name='COUNT')
counts
```

```python
event_types = counts['EVENT_TYPE'].unique()
dictionary = pd.DataFrame(columns=['事件种类'])
dictionary['事件种类'] = event_types
dictionary.to_csv('{}.csv'.format('dictionary_2'),index=False)
```

```python
# 创建一个字典来映射自然灾害事件到等级
severity_levels = {
    'Flood': '中',
    'Fire': '高',
    'Earthquake': '高'
}

# 创建一个新的列来存储灾害等级
counts['灾害等级'] = counts['EVENT_TYPE'].map(severity_levels)
```

```python
# 筛选出ALABAMA州的事件数据
alabama_counts = counts[counts['STATE'] == 'ALABAMA']

# 绘制分布直方图
plt.bar(alabama_counts['EVENT_TYPE'], alabama_counts['COUNT'])
plt.xlabel('事件类型')
plt.ylabel('计数')
plt.title('ALABAMA州事件类型计数分布')
plt.xticks(rotation=90)
plt.show()
```

```python
# 绘制饼状图
plt.pie(alabama_counts['COUNT'], labels=alabama_counts['EVENT_TYPE'], autopct='%1.1f%%',
         startangle=90, wedgeprops={'edgecolor': 'white'})
plt.title('ALABAMA州事件类型计数分布')
# plt.axis('equal')
# 调整布局
plt.tight_layout()

plt.show()
```
